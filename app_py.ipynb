```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Category": ["Food", "Rent", "Transport", "Shopping", "Entertainment", "Savings"],
    "Amount": [5000, 10000, 3000, 4000, 2000, 5000]
}

df = pd.DataFrame(data)
df
plt.figure(figsize=(6,6))
plt.pie(df["Amount"], labels=df["Category"], autopct='%1.1f%%')
plt.title("Monthly Expense Breakdown")
plt.show()
def analyze_finances(df):
    total = df["Amount"].sum()
    savings = df[df["Category"] == "Savings"]["Amount"].values[0]
    rent = df[df["Category"] == "Rent"]["Amount"].values[0]
    food = df[df["Category"] == "Food"]["Amount"].values[0]

    print(f"Total Monthly Expense: ‚Çπ{total}")

    if savings < 0.2 * total:
        print("‚ö†Ô∏è You are saving less than 20% of income. Try to save more.")
    else:
        print("‚úÖ Your savings look healthy.")

    if rent > 0.4 * total:
        print("‚ö†Ô∏è Rent takes more than 40% of your expenses. Consider reducing it.")

    if food > 0.3 * total:
        print("‚ö†Ô∏è Food expenses are high. Track dining out & groceries.")

    print("üí° Tip: Aim for 50% needs, 30% wants, 20% savings (50/30/20 rule).")

analyze_finances(df)
categories = ["Food", "Rent", "Transport", "Shopping", "Entertainment", "Savings"]
amounts = []

for cat in categories:
    amt = int(input(f"Enter monthly expense for {cat}: ‚Çπ"))
    amounts.append(amt)

df = pd.DataFrame({"Category": categories, "Amount": amounts})
print(df)

analyze_finances(df)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.set_page_config(page_title="AI Personal Finance Advisor", layout="wide")

st.title("üí∞ AI Personal Finance Advisor")
st.write("Analyze your monthly expenses and get personalized financial advice!")

# File uploader
uploaded_file = st.file_uploader("Upload your monthly expenses CSV (Category, Amount)", type=["csv"])

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)
else:
    # Sample Data
    data = {
        "Category": ["Food", "Rent", "Transport", "Shopping", "Entertainment", "Savings"],
        "Amount": [5000, 10000, 3000, 4000, 2000, 5000]
    }
    df = pd.DataFrame(data)

st.subheader("üìä Expense Data")
st.dataframe(df)

# Pie chart
fig, ax = plt.subplots()
ax.pie(df["Amount"], labels=df["Category"], autopct='%1.1f%%')
ax.set_title("Monthly Expense Breakdown")
st.pyplot(fig)

# Analysis function
def analyze_finances(df):
    total = df["Amount"].sum()
    savings = df[df["Category"] == "Savings"]["Amount"].values[0] if "Savings" in df["Category"].values else 0

    st.subheader("üß† AI Financial Advice")
    st.write(f"**Total Monthly Expense:** ‚Çπ{total}")

    if savings < 0.2 * total:
        st.warning("‚ö†Ô∏è You are saving less than 20% of income. Try to save more.")
    else:
        st.success("‚úÖ Your savings look healthy.")

    if "Rent" in df["Category"].values:
        rent = df[df["Category"] == "Rent"]["Amount"].values[0]
        if rent > 0.4 * total:
            st.warning("‚ö†Ô∏è Rent takes more than 40% of your expenses. Consider reducing it.")

    if "Food" in df["Category"].values:
        food = df[df["Category"] == "Food"]["Amount"].values[0]
        if food > 0.3 * total:
            st.warning("‚ö†Ô∏è Food expenses are high. Track dining out & groceries.")

    st.info("üí° Tip: Aim for 50% needs, 30% wants, 20% savings (50/30/20 rule).")

analyze_finances(df)
```
